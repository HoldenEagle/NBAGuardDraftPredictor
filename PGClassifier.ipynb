```python
#loading data , making dataframe, and label encoding a feature I got to label encode
import pandas as pd
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

ncaa_df = pd.read_csv("NCAA_Stats_Preprocessed.csv")
pg_df = ncaa_df[
    (ncaa_df['Role/Position'] == 'Scoring PG') | (ncaa_df['Role/Position'] == 'Pure PG')
]

#print(len(pg_df))
pg_df['Role/Position'] = le.fit_transform(pg_df['Role/Position'])
#pg_df = pg_df[
    #(pg_df['pts'] >= 8)
#]
#print(pg_df.head(10))
#print(len(pg_df))
```

```python
#this is how I got the features that I used. I played around with a few
print(pg_df.columns)
sub_ncaa_df = pg_df[['team_encoded','conf_encoded', 'yr_encoded', 'ht_encoded', 'Rec Rank', 'usg', 'FTM','eFG', 'ast', 'stl', 'pts', 'TPM', 'adrtg' ,'Role/Position', 'Drafted_that_year']]
pg_df.head(20)
```

```python
#implement train test split to make a training and testing set
from sklearn.model_selection import train_test_split

X = sub_ncaa_df.drop('Drafted_that_year', axis=1)
y = sub_ncaa_df['Drafted_that_year']

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    stratify=y,      # Keep class distribution balanced
    random_state=42
)
```

```python
#import xgboost library
import xgboost as xgb
from sklearn.metrics import accuracy_score, classification_report
```

```python
#implement our best model and run it on the testing set. Add weights to the drafted class 
#to increase performance
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

model = xgb.XGBClassifier(use_label_encoder=False, eval_metric=['auc', 'logloss'])

scale_pos_weight = len(y_train) / (4.5 * sum(y_train))

model =xgb.XGBClassifier(
    eval_metric=['auc', 'logloss'],
    learning_rate=0.1,
    max_depth=4,
    min_child_weight=3,
    n_estimators=300,
    scale_pos_weight= scale_pos_weight,
    use_label_encoder=False
)

# Fit the model
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
