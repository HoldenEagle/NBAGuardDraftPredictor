```python
#loading data , making dataframe, and label encoding a feature I got to label encode
import pandas as pd
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

ncaa_df = pd.read_csv("NCAA_Stats_Preprocessed.csv")

pg_df = ncaa_df[
    ((ncaa_df['Role/Position'] == 'Scoring PG') | (ncaa_df['Role/Position'] == 'Pure PG')) & (ncaa_df['Drafted_that_year'] == 1)
]

#print(len(pg_df))
pg_df['Role/Position'] = le.fit_transform(pg_df['Role/Position'])

pg_df = pg_df[pg_df['player_name'] != 'James Johnson']
print(len(pg_df))
pg_df.head(20)
```

```python
print(pg_df['pick'])
```

```python
#this is how I got the features that I used. I played around with a few
print(pg_df.columns)
sub_ncaa_df = pg_df[['player_name', 'team_encoded','ht_encoded','conf_encoded', 'yr_encoded','rimmade/(rimmade+rimmiss)', 'midmade', 'pfr', 'Rec Rank', 'obpm', 'ast/tov', 'FTM', 'TO_per', 'ast', 'pts' , 'adrtg','Role/Position_encoded', 'pick']]
sub_ncaa_df.head(20)
```

```python
#drop null rows
sub_ncaa_df = sub_ncaa_df.dropna()
```

```python
#implement train test split to make a training and testing set
from sklearn.model_selection import train_test_split

X = sub_ncaa_df.drop(['pick' , 'player_name'], axis=1)
y = sub_ncaa_df['pick']
indices = X.index

X_train, X_test, y_train, y_test,idx_train, idx_test = train_test_split(
    X, y, indices,
    test_size=0.2,
    random_state=42
)
print(X_train.dtypes)
print(idx_test)
```

```python
#looking at best model and testing it on the testing set
from sklearn.preprocessing import MinMaxScaler
import xgboost as xgb
import time
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
model = xgb.XGBRegressor(
    objective='reg:logistic',
    booster='dart',
    colsample_bytree=0.9,
    learning_rate=0.07,
    max_depth=5,
    min_child_weight=1,
    n_estimators=265,
    random_state=42,
    subsample=1
)
scaler = MinMaxScaler()
y_train = np.array(y_train).reshape(-1, 1)
y_train_scaled = scaler.fit_transform(y_train).ravel()

# Train
model.fit(X_train, y_train_scaled)

# Predict
y_pred_scaled = model.predict(X_test)

y_pred = scaler.inverse_transform(y_pred_scaled.reshape(-1, 1)).ravel()

# Evaluate
mse = mean_squared_error(y_test, y_pred)


rmse = np.sqrt(mse)
print(f"Root Mean Squared Error: {rmse:.4f}")

for i in idx_test:
    row = ncaa_df.iloc[i]
    actual_pick = row['pick']
    player_data = row[['team_encoded', 'conf_encoded','ht_encoded', 'yr_encoded','rimmade/(rimmade+rimmiss)' , 'midmade',
                       'pfr', 'Rec Rank', 'obpm',
                       'ast/tov', 'FTM', 'TO_per', 'ast', 'pts' , 'adrtg' ,'Role/Position_encoded']].values.reshape(1, -1)
    predicted_pick = model.predict(player_data)[0]
    predicted_pick = (predicted_pick * 59)+1

    print(f"{row['player_name']} : predicted pick {predicted_pick} , actual pick {actual_pick}")
    time.sleep(1)
```
